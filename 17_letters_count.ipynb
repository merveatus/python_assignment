```python
sentence = input("Please enter a sentence: ")

def char_frequency(sentence):
    dict = {}
    for n in sentence:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
print(char_frequency(sentence))



def word_count(str):
    count = dict()
    words = str.split()

    for word in words:
        if word in count:
            count[word] += 1
        else:
            count[word] = 1

    return count

freq_words = word_count(sentence)
print(freq_words)
```
